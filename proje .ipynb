```python
arasinav = float(input("Ara Sınav Notunuzu Giriniz: "))
final = float(input("Final Notunuzu Giriniz: "))
proje = float(input("Proje Notunuzu Giriniz: "))

ortalamanot = (0.3*arasinav) + (0.5*final) + (0.2*proje)
print("Not ortalamanız: ", ortalamanot)

if (ortalamanot >= 90):
    print("Harf notunuz: AA")
elif (ortalamanot >= 70 and ortalamanot <90):
    print("Harf notunuz: BB")
elif (ortalamanot >= 50 and ortalamanot <70):
    print("Harf notunuz: CC")
elif (ortalamanot >= 30 and ortalamanot <50):
    print("Harf notunuz: DD")
    
else:
    print("Harf notunuz FF. Kaldınız!!!")
```
